```python
import pandas as pd
df=pd.read_csv('https://raw.githubusercontent.com/2k0v11/Dataset/main/IPL%20IMB381IPL2013.csv')
df.head(10)
```

```python
X_features = ['AGE', 'COUNTRY', 'PLAYING ROLE', 'T-RUNS', 'T-WKTS', 'ODI-RUNS-S', 'ODI-SR-B', 'ODI-WKTS', 'ODI-SR-BL','CAPTAINCY EXP', 
'RUNS-S', 'HS', 'AVE', 'SR-B','SIXERS', 'RUNS-C', 'WKTS', 'AVE-BL', 'ECON', 'SR-BL']
categorical_feature= ['AGE', 'COUNTRY','PLAYING ROLE','CAPTAINCY EXP']
```

```python
#encoding categorical features 
df_encoded=pd.get_dummies(df[X_features],columns=categorical_feature,drop_first=True)
df_encoded
```

```python
import statsmodels.api as sm
x=sm.add_constant(df_encoded)
y=df['SOLD PRICE']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
from sklearn import linear_model
model=linear_model.LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import r2_score,mean_squared_error
import numpy as np
print('R^2 score:',np.abs(r2_score(y_pred,y_test)))
print("Mean Squared Error:",np.sqrt(mean_squared_error(y_pred,y_test)))
```
