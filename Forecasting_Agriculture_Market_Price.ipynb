#                         Forecasting Agriculture Market Price




```python
import pandas as pd
df=pd.read_csv('agriculture.csv')
df
```

```python
#check for null values
df.isnull().sum()
```

```python
#information of the dataframe
df.info()
```

```python
#choosing input and output variable
# I have chosen Date and Modal Price
```

```python
df.rename(columns={'Price Date':'ds','Modal Price (Rs./Quintal)':'y'},inplace=True) # inplace= True is given to avoid duplications
```

```python
df
```

```python
new_df=df[['ds','y']]
```

```python
new_df
```

```python
new_df.info()
```

```python
# The columns x and y are not in required datatypes, so we have to convert
#we convert x to datetime format
#we convert y to integer format
```

```python
new_df['ds']=pd.to_datetime(new_df['ds'])
new_df
```

```python

new_df['y']= new_df['y'].astype(int)
new_df
```

```python
new_df.info() #datatypes of both x and y coverted to required format
```

```python
#visualisation
from matplotlib import pyplot as plt
plt.figure(figsize=(25,5))
plt.scatter(new_df['ds'],new_df['y'])
```

```python
#!conda install -c conda-forge fbprophet -y
%pip install prophet
```

```python
#FB PROPHET
from prophet import Prophet
model=Prophet()
model.fit(new_df)
```

```python
forecast=model.make_future_dataframe(periods=365, freq='D')
forecast=model.predict(forecast)
forecast
```

```python
model.plot(forecast,xlabel='Date',ylabel='Potato Prices/Quintal')
```
