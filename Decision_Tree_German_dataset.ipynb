**GERMAN CREDIT DATA**

```python
import pandas as pd
df=pd.read_csv('https://raw.githubusercontent.com/2k0v11/Dataset/main/German%20Credit%20Data.csv')
df
```

```python
df.iloc[:5,:8]
```

```python
df.iloc[:5,8:]
```

```python
df['savings_acc'].unique()
```

```python
x_dummies=list(df.columns)
x_dummies.remove('status',)
encoded_df=pd.get_dummies(df[x_dummies],drop_first=True)
```

```python
x=encoded_df
y=df['status']
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=42,test_size=0.3)
```

```python
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier(criterion='gini',max_depth=4)
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
from sklearn import metrics
print('Confusion matrix:',metrics.confusion_matrix(y_pred,y_test))
print('Accuracy Score:',metrics.accuracy_score(y_pred,y_test))
```

```python
from sklearn.tree import export_graphviz
import pydotplus as pdot
from IPython.display import Image
export_graphviz(model,out_file = "tree.odt", feature_names = x_train.columns, filled = True)
graph = pdot.graphviz.graph_from_dot_file("tree.odt")
graph.write_jpg("tree.png")
Image(filename = "tree.png")
```
